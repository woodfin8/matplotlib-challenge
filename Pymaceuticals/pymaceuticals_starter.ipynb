```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse = pd.read_csv(mouse_drug_data_to_load)
trial = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
dataset = pd.merge(trial,mouse,on="Mouse ID")

#mouse.head()
# Display the data table for preview
dataset.head()
```

#### Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint \
drug_mean = dataset.groupby(["Drug","Timepoint"],as_index=False)[["Tumor Volume (mm3)"]].mean()

drug_mean.head()



# Convert to DataFrame

# Preview DataFrame
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
#dataset["Tumor Volume (mm3)"] = dataset["Tumor Volume (mm3)"].astype(float)
drug_sem = dataset.groupby(["Drug","Timepoint"])[["Tumor Volume (mm3)"]].sem().reset_index()
# Convert to DataFrame
drug_sem.head()
# Preview DataFrame

```

```python
# Minor Data Munging to Re-Format the Data Frames
drug_pivot = drug_mean.pivot_table(index='Timepoint', columns='Drug')["Tumor Volume (mm3)"]
sem_pivot = drug_sem.pivot_table(index="Timepoint",columns="Drug")["Tumor Volume (mm3)"]
# Preview that Reformatting worked
drug_pivot


```

```python
# Generate the Plot (with Error Bars)
fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(20,8))
ax.errorbar(drug_pivot.index,drug_pivot["Capomulin"], yerr = sem_pivot["Capomulin"],fmt="o-", color="r", alpha = 0.5, label="Capomulin")
ax.errorbar(drug_pivot.index,drug_pivot["Ceftamin"], yerr = sem_pivot["Ceftamin"],fmt="v-", color="y", alpha = 0.5, label="Ceftamin")
ax.errorbar(drug_pivot.index,drug_pivot["Infubinol"], yerr = sem_pivot["Infubinol"],fmt="s-", color="b", alpha = 0.5, label="Infubinol")
ax.errorbar(drug_pivot.index,drug_pivot["Ketapril"], yerr = sem_pivot["Ketapril"],fmt="p-", color="violet", alpha = 0.5, label="Ketapril")
ax.errorbar(drug_pivot.index,drug_pivot["Placebo"], yerr = sem_pivot["Placebo"],fmt="+-", color="black", alpha = 0.5, label="Placebo")
ax2.errorbar(drug_pivot.index,drug_pivot["Placebo"], yerr = sem_pivot["Placebo"],fmt="+-", color="black", alpha = 0.5, label="Placebo")
ax2.errorbar(drug_pivot.index,drug_pivot["Naftisol"], yerr = sem_pivot["Naftisol"],fmt="x-", color="orange", alpha = 0.5, label="Naftisol")
ax2.errorbar(drug_pivot.index,drug_pivot["Propriva"], yerr = sem_pivot["Propriva"],fmt="d-", color="g", alpha = 0.5, label="Propriva")
ax2.errorbar(drug_pivot.index,drug_pivot["Ramicane"], yerr = sem_pivot["Ramicane"],fmt="^-", color="brown", alpha = 0.5, label="Ramicane")
ax2.errorbar(drug_pivot.index,drug_pivot["Stelasyn"], yerr = sem_pivot["Stelasyn"],fmt="*-", color="purple", alpha = 0.5, label="Stelasyn")

ax.set_title("Tumor Response to Treatment (Group 1)", size=16)
ax2.set_title("Tumor Response to Treatment (Group 2)",size=16)
ax.set_xlabel("Timepoint")
ax.set_ylabel("Tumor Volume (mm3)")
ax2.set_xlabel("Timepoint")
ax2.set_ylabel("Tumor Volume (mm3)")
# plt.legend(loc="best", fontsize="small", fancybox=True)
ax.yaxis.grid()
ax2.yaxis.grid()
ax.legend(loc="best", fontsize = "large", fancybox=True)
ax2.legend(loc="best", fontsize = "large", fancybox=True)


# Save the Figure

plt.savefig("tumor_resopnse.png")


```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_count = dataset.groupby(["Drug","Timepoint"])[["Mouse ID"]].nunique().reset_index()
mouse_count.rename(columns={"Mouse ID":"Mouse Count"}, inplace=True)
# Convert to DataFrame

# Preview DataFrame
mouse_count.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_pivot = mouse_count.pivot_table(index='Timepoint', columns='Drug')["Mouse Count"]
# Preview the Data Frame
mouse_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)


#create percentage columns
for column in mouse_pivot.columns:
    mouse_pivot[column+"%"]=mouse_pivot[column]/mouse_pivot[column][0]
    
    
fig, (ax,ax2) = plt.subplots(ncols=2,figsize=(20,8))
ax.errorbar(mouse_pivot.index,mouse_pivot["Capomulin%"], fmt="o-", color="r", alpha = 0.5, label="Capomulin")
ax.errorbar(mouse_pivot.index,mouse_pivot["Ceftamin%"],fmt="v-", color="y", alpha = 0.5, label="Ceftamin")
ax.errorbar(mouse_pivot.index,mouse_pivot["Infubinol%"],fmt="s-", color="b", alpha = 0.5, label="Infubinol")
ax.errorbar(mouse_pivot.index,mouse_pivot["Ketapril%"],fmt="p-", color="violet", alpha = 0.5, label="Ketapril")
ax.errorbar(mouse_pivot.index,mouse_pivot["Placebo%"],fmt="+-", color="black", alpha = 0.5, label="Placebo")
ax2.errorbar(mouse_pivot.index,mouse_pivot["Placebo%"],fmt="+-", color="black", alpha = 0.5, label="Placebo")
ax2.errorbar(mouse_pivot.index,mouse_pivot["Naftisol%"],fmt="x-", color="orange", alpha = 0.5, label="Naftisol")
ax2.errorbar(mouse_pivot.index,mouse_pivot["Propriva%"], fmt="d-", color="g", alpha = 0.5, label="Propriva")
ax2.errorbar(mouse_pivot.index,mouse_pivot["Ramicane%"], fmt="^-", color="brown", alpha = 0.5, label="Ramicane")
ax2.errorbar(mouse_pivot.index,mouse_pivot["Stelasyn%"], fmt="*-", color="purple", alpha = 0.5, label="Stelasyn")
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
vals2 = ax2.get_yticks()
ax2.set_yticklabels(['{:,.0%}'.format(x) for x in vals2])

ax.set_title("Survival During Treatment (Group 1)", size=16)
ax2.set_title("Survival During Treatment (Group 2)",size=16)
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Survival Rate (%)")
ax2.set_xlabel("Time (Days)")
ax2.set_ylabel("Survival Rate (%)")
ax.yaxis.grid()
ax2.yaxis.grid()
ax.legend(loc="best", fontsize = "large", fancybox=True)
ax2.legend(loc="best", fontsize = "large", fancybox=True)

plt.savefig("survival.png")

plt.show()

    
```
